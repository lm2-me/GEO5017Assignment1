```python
import numpy as np
import matplotlib.pyplot as plt

# plt.style.use('_mpl-gallery')


pointcloudsdummy = np.random.randn(800).reshape((100,8))

#pointcloudsdummy = np.random.randint(0,500,3000).reshape((500,6))


D = np.transpose(pointcloudsdummy)

# plot:
fig, ax = plt.subplots()

vp = ax.violinplot(D, widths=1,
                   showmeans=False, showmedians=False, showextrema=False)
# styling:
for body in vp['bodies']:
    body.set_alpha(0.9)
ax.set(xlim=(0, 9), xticks=np.arange(1, 9),
       ylim=(0, 2), yticks=np.arange(1, 2))

plt.show()
```
